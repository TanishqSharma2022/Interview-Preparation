```python
from pyspark.sql import SparkSession
import pyspark
```

```python
spark = SparkSession.builder.appName("Data Cleaning").master("local").getOrCreate()
```

```python
data = spark.read.csv("house.csv", header=True, inferSchema=True)
# df = spark.createDataFrame(data, columns)
# df.show()
```

```python
data.show()
```

```python
data.printSchema()
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StructType, StructField, IntegerType, DoubleType
```

```python
data.select("Home", "SqFt", "Price").show()
```

```python
# data.filter(data.SqFt > 1000).show()
data.filter(col("SqFt") > 1000).show()
```

```python
data.filter((data.SqFt > 1000) & (data.Bedrooms < 5)).show()
```

```python
from pyspark.sql.functions import when

data = data.withColumn("High Price", 
                when(col("Price") > 100000, "High")
                .otherwise("Low")
                )
data.show()
```
