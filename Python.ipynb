```python
class Employee:
    raise_amount = 1.04 # class variable can be accessed by any instance, is not changed 

    def __init__(self, first, last, pay): # self is nothing but the instance. 
        self.first = first # these are instance variables
        self.last = last
        self.pay = pay

    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    @classmethod
    def change_raise_amount(cls, amt): # class method takes class as the first arguments instead of instance
        cls.raise_amount = amt

emp1 = Employee("Tanishq", "Sharma", 50000)
emp1.raise_amount
Employee.change_raise_amount(1.9) # we can use both class and instance to use this class method 
emp1.change_raise_amount(1.9)
emp1.raise_amount
```

### `Decorators`

```python
def decorator(func):
    def wrapper():
        print("Before Function ")
        func()
        print("After Function ")

    return wrapper
```

```python
@decorator
def greet():
    print("Hello")

greet()
```

```python
decorator(greet)
```

```python
greet.__name__
```
